```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import multi_loc.covariance as covariance
```

```python
grid_points = 10
x = np.arange(grid_points)
x, y = np.meshgrid(x, x)
# x = np.minimum(x, (grid_points - x) % grid_points)
# y = np.minimum(y, (grid_points - y) % grid_points)
rho = np.sqrt(x**2 + y**2)
rho0 = 100
```

```python
plt.figure()
image = plt.imshow(rho)
plt.colorbar(image)
plt.title('distance from (0, 0)')
```

```python
rho = rho.ravel()
P_rho = covariance.make_correlation_matrix(
    rho, rho0, covariance.correlation_sqd_exp)
```

```python
plt.figure()
image = plt.imshow(P_rho)
plt.colorbar(image)
plt.title('Covariance Function for rho')
```
